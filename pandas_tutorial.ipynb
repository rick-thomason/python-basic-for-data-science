```python
netflix['computer_date'] = pd.to_datetime(netflix['date_added'])
```

```python
netflix['computer_date'].dropna()
```

```python
from datetime import datetime as dt
```

```python
netflix['computer_date'].fillna(dt(2020, 1, 1), inplace=True)
```

```python
netflix
```

```python
netflix['unix_time'] = netflix['computer_date'].view(int)
```

```python
netflix
```

```python
netflix['date_added'].str.split(', ', expand=True)
```

```python
netflix[['Date Part 1', 'Date Part 2']] = netflix['date_added'].str.split(', ', expand=True)
netflix
```

```python
netflix.rename(columns={
    'Date Year': 'Year'
}, inplace=True)
```

```python
netflix['cast'].fillna(value='no cast', inplace=True)
```

```python
netflix['genre_count'] = netflix['listed_in'].map(lambda x: len(x.split(',')))
```

```python
unique_genres = netflix['listed_in'].unique()
```

```python
netflix_movies_by_country = netflix.groupby('country')['mshow_id'].count().reset_index()
```

```python
netflix_movies_by_country.rename(columns={
    'mshow_id': 'show_id'
}, inplace=True)
```

```python
netflix_movies_by_country.sort_values(by='show_id', ascending=False)
```
